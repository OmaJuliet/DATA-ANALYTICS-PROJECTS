```python
# Load and Inspect the Data

import pandas as pd

# Load the CSV file
# specify a different encoding when reading the file: ISO-8859-1, which often works with CSV files that have encoding issues
df = pd.read_csv('world_population.csv', encoding='ISO-8859-1')

# View basic info and the first few rows of the dataset
df.info()
df.head()
```

```python
# Handle Missing Data

# Check for missing values
df.isnull().sum()
```

```python
# Display basic statistics for numerical columns
df.describe()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Standardize column names
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.info()
```

```python
# Group by continent to calculate total population in 2023
population_by_continent = df.groupby('continent')['2023_population'].sum()

# View the result
population_by_continent
```

```python
# Sort countries by their population in 2023
df_sorted_growth = df[['country/territory', '2023_population']].sort_values(by='2023_population', ascending=False)

# Top 10 countries by population in the year 2023
df_sorted_growth.head(10)
```

```python
# Calculate population growth between 2000 and 2023
df['growth_2000_to_2023'] = (df['2023_population'] - df['2000_population']) / df['2000_population'] * 100
```

```python
df.info()
df.head()
```

```python
# Sort countries by their population growth between 2000 and 2023
df_sorted_growth = df[['country/territory', 'growth_2000_to_2023']].sort_values(by='growth_2000_to_2023', ascending=False)

# Top 10 countries by population growth
df_sorted_growth.head(10)
```

```python
# Sort countries by their population growth between 2000 and 2023
df_sorted_growth = df[['country/territory', 'growth_2000_to_2023']].sort_values(by='growth_2000_to_2023', ascending=True)

# Bottom 10 countries by population growth
df_sorted_growth.head(10)
```

```python
# Calculate population growth rates over the last decade
df['growth_2010_to_2020'] = (df['2020_population'] - df['2010_population']) / df['2010_population'] * 100

df.info()
df.head()
```

```python
# Sort countries by their population growth between 2010 and 2020
df_sorted_growth = df[['country/territory', 'growth_2010_to_2020']].sort_values(by='growth_2010_to_2020', ascending=False)

# Top 10 countries by population growth between 2010 and 2020
df_sorted_growth.head(10)
```

```python
# Sort countries by their population growth between 2000 and 2023
df_sorted_growth = df[['country/territory', 'growth_2010_to_2020']].sort_values(by='growth_2010_to_2020', ascending=True)

# Bottom 10 countries by population growth
df_sorted_growth.head(10)
```

```python
# DATA VISUALIZATION
import matplotlib.pyplot as plt
import seaborn as sns

# Population Distribution by Continent (Bar Plot)

# Bar plot of total population by continent
plt.figure(figsize=(10, 6))
sns.barplot(x=population_by_continent.index, y=population_by_continent.values)
plt.title('Total Population by Continent (2023)')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
# Population Growth Over Time for Selected Countries

# Select key countries for a population trend comparison
countries = ['China', 'India', 'United States', 'Brazil', 'Nigeria', 'United Kingdom', 'Canada', 'South Africa', 'Australia']

# Plot population growth over time for selected countries
plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot([1970, 1980, 1990, 2000, 2010, 2020, 2023], 
             df[df['country/territory'] == country][['1970_population', '1980_population', '1990_population', 
                                                    '2000_population', '2010_population', '2020_population', 
                                                    '2023_population']].values.flatten(), label=country)

plt.title('Population Growth Over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Population Growth Over Time for for the Top 10 Populated Countries

# Select top 10 populated countries for a population trend comparison
countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan', 'Nigeria', 'Brazil', 'Bangladesh', 'Russia', 'Mexico']

# Plot population growth over time for the top 10 populated countries
plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot([1970, 1980, 1990, 2000, 2010, 2020, 2023], 
             df[df['country/territory'] == country][['1970_population', '1980_population', '1990_population', 
                                                    '2000_population', '2010_population', '2020_population', 
                                                    '2023_population']].values.flatten(), label=country)

plt.title('Population Growth Over Time for the Top 10 Most Populated Countries')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Population Density Distribution (Histogram)

# Histogram of population density
plt.figure(figsize=(8, 6))
sns.histplot(df['density_(kmâ²)'], bins=30, kde=True)
plt.title('Population Density Distribution')
plt.xlabel('Density (per km²)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Projected Population Growth Rate for Each Continent by Year 2030

# Project population for 2030 using compound annual growth rate--CAGR (assuming constant growth)
df['projected_growth_rate'] = (df['2023_population'] / df['2000_population']) ** (1 / 23) - 1
df['projected_population_2030'] = df['2023_population'] * (1 + df['projected_growth_rate']) ** 7

# Group by continent and sum projected populations for 2030
projected_by_continent = df.groupby('continent')['projected_population_2030'].sum().sort_values()

# Plot projected population by continent
plt.figure(figsize=(10, 6))
sns.barplot(x=projected_by_continent.index, y=projected_by_continent.values)
plt.title('Projected Population by Continent in 2030')
plt.xlabel('Continent')
plt.ylabel('Projected Population (in billions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Countries by Population Percentage

# Sort countries by world population percentage
top_10_countries = df[['country/territory', 'world_percentage']].sort_values(by='world_percentage', ascending=False).head(10)

# Plot the top 10 contributors to the world population
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_countries['country/territory'], y=top_10_countries['world_percentage'])
plt.title('Top 10 Countries Contributing to World Population (2023)')
plt.xlabel('Country')
plt.ylabel('World Population Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify Countries with Highest and Lowest Population Density

# Sort countries by population density
highest_density = df[['country/territory', 'density_(kmâ²)']].sort_values(by='density_(kmâ²)', ascending=False).head(10)
lowest_density = df[['country/territory', 'density_(kmâ²)']].sort_values(by='density_(kmâ²)', ascending=True).head(10)

# Plot the countries with highest population density
plt.figure(figsize=(10, 6))
sns.barplot(x=highest_density['country/territory'], y=highest_density['density_(kmâ²)'])
plt.title('Top 10 Countries with Highest Population Density')
plt.xlabel('Country')
plt.ylabel('Population Density (people per km²)')
plt.xticks(rotation=45)
plt.show()

# Similarly for lowest density
plt.figure(figsize=(10, 6))
sns.barplot(x=lowest_density['country/territory'], y=lowest_density['density_(kmâ²)'])
plt.title('Top 10 Countries with Lowest Population Density')
plt.xlabel('Country')
plt.ylabel('Population Density (people per km²)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Population Density Changes Over Time in Continents

# Calculate population density over time (for example, for 1970, 2000, and 2023)
df['density_1970'] = df['1970_population'] / df['area_(km_)']
df['density_2000'] = df['2000_population'] / df['area_(km_)']
df['density_2023'] = df['2023_population'] / df['area_(km_)']

# Group by continent and calculate the mean density
density_by_continent = df.groupby('continent')[['density_1970', 'density_2000', 'density_2023']].mean()

# Plot density over time for each continent
density_by_continent.plot(kind='bar', figsize=(12, 6))
plt.title('Population Density Change Over Time by Continent')
plt.xlabel('Continent')
plt.ylabel('Density (people per km²)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Growth Rates in Developed vs. Developing Countries


# Create a column to categorize countries
developed_countries = ['United States', 'Germany', 'China', 'Japan', 'Australia', 'Canada']
df['development_status'] = df['country/territory'].apply(lambda x: 'Developed' if x in developed_countries else 'Developing')

# Compare growth rates between developed and developing countries
growth_by_status = df.groupby('development_status')['growth_2000_to_2023'].mean()

# Plot growth rates
plt.figure(figsize=(8, 6))
sns.barplot(x=growth_by_status.index, y=growth_by_status.values)
plt.title('Population Growth Rate in Developed vs. Developing Countries (2000-2023)')
plt.xlabel('Development Status')
plt.ylabel('Growth Rate (%)')
plt.show()
```

```python
# Population Density vs. Area

# Calculate correlation between density and area
correlation = df[['density_(kmâ²)', 'area_(km_)']].corr()

# Plot correlation
sns.scatterplot(x='area_(km_)', y='density_(kmâ²)', data=df)
plt.title('Population Density vs. Country Area')
plt.xlabel('Area (km²)')
plt.ylabel('Density (people per km²)')
plt.show()

# Display correlation value
print(correlation)
```

```python
# Countries Under Most Population Pressure (High Density and High Growth Rate)

# Filter countries with both high density and high growth rates
high_pressure_countries = df[(df['density_(kmâ²)'] > df['density_(kmâ²)'].quantile(0.75)) & (df['growth_2000_to_2023'] > df['growth_2000_to_2023'].quantile(0.75))]

# Plot these countries
plt.figure(figsize=(10, 6))
sns.scatterplot(x='density_(kmâ²)', y='growth_2000_to_2023', data=high_pressure_countries, hue='country/territory', s=100)
plt.title('Countries with High Population Pressure (High Density and High Growth)')
plt.xlabel('Population Density (people per km²)')
plt.ylabel('Population Growth Rate (%)')
plt.show()
```
